**EV Growth Vs Charging Infrastructure in India: A Data-Driven Analysis**

### Project Overview
This project analyzes the relationship between electric vehicle (EV) adoption and charging infrastructure growth across Indian states (2020–2024).
The objective is to highlight regional imbalances and identify states where infrastructure lags behind adoption.
By combining EV count data with charging station data, the analysis provides actionable insights for policymakers, businesses, and sustainability initiatives.


*OBJECTIVE OF THE PROJECT*

To analyze the relationship between electric vehicle (EV) growth and charging infrastructure development across Indian states, highlighting regional imbalances and identifying areas where infrastructure is lagging behind adoption.

*Total EVs per year*

```python
df_comb.groupby('year')['ev_count'].sum()
```

*Total Charging Stations Per Year*

```python
df_comb.groupby('year')['station_count'].sum()
```

*EV Vs Stations*

```python
ratio = df_comb.groupby('state').apply( lambda x: x['ev_count'].sum() / x['station_count'].sum())
```

```python
print(ratio.sort_values)
```

*Top states by EV count*

```python
top_states_ev=df_comb.groupby('state')['ev_count'].sum().sort_values(ascending=False)
```

```python
top_states_ev
```

*Top states by Station count*

```python
top_states_st= df_comb.groupby('state')['station_count'].sum().sort_values(ascending=False)
```

```python
top_states_st
```

# Data Visualization

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
sns.lineplot(data=df_comb, x= 'year', y='ev_count', hue='state')
plt.title('EV Growth Over Years By State')
plt.show()
```

```python
sns.lineplot(data=df_comb, x= 'year', y='station_count', hue='state')
plt.title('Charging Station Growth Over Years By State')
plt.show()
```

```python
sns.scatterplot( data=df_comb, x='ev_count', y='station_count', hue= 'state')
plt.title('EV Vs Charging Stations')
plt.show()
```

*KEY OBSERVATIONS*

EV Growth by State

Maharashtra leads with the highest electric vehicle growth over the years.

Gujarat shows the lowest EV growth among the states analyzed.

Charging Infrastructure Growth by State

Delhi has the fastest growth in EV charging stations.

Maharashtra ranks second in charging infrastructure expansion.

Gujarat again ranks lowest in charging station growth.

**CONCLUSION**

There is a clear imbalance between EV adoption and charging infrastructure across states. While Maharashtra is rapidly adopting EVs, Delhi is leading in building the necessary support systems. Gujarat, on the other hand, lags behind in both EV growth and infrastructure, which could slow down future adoption.

*SUGGESTION*

To support sustainable EV growth, states like Gujarat need targeted investment and policy support to expand charging infrastructure. A balanced approach—where EV adoption and infrastructure grow together—is essential to avoid bottlenecks and ensure user confidence. Collaboration between government, private players, and local communities can accelerate this transition.
